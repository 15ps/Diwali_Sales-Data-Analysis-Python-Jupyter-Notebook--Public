# Step 4 : EDA

Gender

```python
df.columns
```

```python
gender=sns.countplot(x='Gender',data=df)
for bars in gender.containers:
    gender.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
```

```python
sales_by_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount',data=sales_by_gender)
```

```python
#conclusion:most of buyers are females. Purchasing power of females are greater then males
```

Age

```python
sns.countplot(x='Age Group',data=df)
```

```python
sns.countplot(x='Age Group',hue='Gender',data=df)
```

```python
age_group=sns.countplot(data=df,x='Age Group',hue="Gender")
for bars in age_group.containers:
    age_group.bar_label(bars)
```

```python
#conclusion :36-35 age group women buys most
```

```python
#total amount vs age group
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount')
```

```python
sales_by_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount',data=sales_by_age)
```

```python
#conclusion:most of buyers are of agegroup 26-35 who are max females
```

state

```python
df.columns
```

```python
#to find top 10 total orders from states
```

```python
Order_by_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='State', y='Orders',data=Order_by_state)
```

```python
#to find top 10 total amount/sales from states
```

```python
Order_by_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='State', y='Amount',data=Order_by_state)
```

```python
#for both orders & amount-UP,MAHA,Karnataka are top 3 states
```

Marital Status

```python
MS=sns.countplot(x='Marital_Status',data=df)
plt.figure(figsize=(10,5))
for bars in MS.containers:
    MS.bar_label(bars)
```

```python
#check marital status with amount
```

```python
gender_by_maritalstatus = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x='Marital_Status',y='Amount',data=gender_by_maritalstatus,hue='Gender')
```

```python
#unmarried women buys more then other categories
```

Occupation

```python
plt.figure(figsize=(18,5))
occupation=sns.countplot(data=df,x='Occupation')
for bars in occupation.containers:
    occupation.bar_label(bars)
```

```python
#amount by occupation
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x='Occupation', y='Amount',data=sales_occupation)
```

```python
#conclusion:from above graphs we can see that most of the buyers are working in IT,Aviation,Healthcare sector
```

Product catogory

```python
plt.figure(figsize=(25,5))
Product_Category=sns.countplot(data=df,x='Product_Category')
for bars in Product_Category.containers:
    Product_Category.bar_label(bars)
```

```python
#conclusion:in terms of count of orders happened in clothing,food,electronics
```

```python
#amount by product category
```

```python
sales_productgroup=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(30,5))
sns.barplot(x='Product_Category', y='Amount',data=sales_productgroup)
```

```python
#conclusion:in terms of value/sales/amount of orders happened in clothing,electronics,food
```

```python
#top selling products using product id
```

```python
topselling_productid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(20,5))
sns.barplot(x='Product_ID', y='Orders',data=topselling_productid)
```
